```python
#DISTRIBUTION P
import seaborn as sns
%matplotlib inline  
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=30)
```

```python
sns.jointplot(x='total_bill' , y='tip' , data = tips, kind='hex')
```

```python
sns.pairplot(tips,hue='sex', diag_kind='hist')
```
