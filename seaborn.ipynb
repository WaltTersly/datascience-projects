```python
#DISTRIBUTION P
import seaborn as sns
%matplotlib inline  
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=30)
```

```python
sns.jointplot(x='total_bill' , y='tip' , data = tips, kind='hex')
```

```python
sns.pairplot(tips,hue='sex', diag_kind='hist')
```

```python
import seaborn as sns
%matplotlib inline  
iris = sns.load_dataset('iris')
iris.head()

```

```python
g = sns.pairplot(iris )
```

```python
sns.PairGrid(iris)
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_lower(sns.scatterplot)
g.map_upper(sns.kdeplot)
```

```python
import seaborn as sns
%matplotlib inline  
tips = sns.load_dataset('tips')
tips.head()
```

```python
gr = sns.FacetGrid(data=tips, col='time' , row='smoker')
gr.map(sns.distplot, 'total_bill')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt 
tips = sns.load_dataset('tips')
tips.head()

gr = sns.FacetGrid(data=tips, col='time' , row='smoker')
gr.map(plt.scatter, 'total_bill','tip')
```

```python
#REGERESSION PLOTS

import seaborn as sns
import matplotlib.pyplot as plt 
tips = sns.load_dataset('tips')
tips.head()

sns.lmplot(x = 'total_bill', y = 'tip', data=tips, hue='day', size=8)
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt
%matplotlib inline 
sns.set_style('whitegrid')
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
sns.jointplot(x='fare', y='age',data=titanic)
```

```python
sns.distplot(titanic['fare'],kde = False, color = "red", bins = 30)
```

```python
sns.boxplot(x='class', y='age', data=titanic, palette='rainbow')
```

```python
sns.swarmplot(x='class', y='age', data=titanic, palette='Set2')
```

```python
sns.countplot(x='sex', data=titanic)
```

```python
sns.heatmap(titanic.corr(), cmap='coolwarm')
plt.title('titanic.corr')
```

```python
grd = sns.FacetGrid(data=titanic, col='sex')
grd.map(plt.hist, 'age')
```
