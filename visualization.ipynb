```python
import numpy as np 
x = np.linspace(0,5,11)
y = x ** 2

x
```

```python
y
```

```python
import matplotlib.pyplot as plt

#functional method
plt.plot(x,y,'r-')
plt.xlabel('X LABEL')
plt.ylabel('Y LABEL')
plt.title('Title')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y, 'b')

plt.subplot(1,2,2)
plt.plot(y,x, 'r')
```

```python
#object oriented method
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X LABEL')
axes.set_ylabel('Y LABEL')
axes.set_title('oop plot')
```

```python
axes
```

```python
#figure size
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.3,0.5,0.4,0.3])
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(10,2),dpi=100)

axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline  
import numpy as np 
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
fig = plt.figure()
ax =  fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel('values of x')
ax.set_ylabel('x by 2')
ax.set_title('x against x by 2')
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.2,.2])
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.2,.2])
ax1.plot(x,y, color = 'red')
ax1.set_xlabel('values of x')
ax1.set_ylabel('x by 2')
ax2.plot(x,y, color = 'red')
ax2.set_xlabel('values of x')
ax2.set_ylabel('x by 2')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.4,.4])
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.4,.4])
ax.plot(x,z, color = 'red')
ax.set_xlabel('values of x')
ax.set_ylabel('x squared')

ax2.plot(x,y, color = 'red')
ax2.set_xlabel('values of x')
ax2.set_ylabel('x by 2')
ax2.set_ylim([30,50])
ax2.set_xlim([20,22])
```

```python
fig,axes = plt.subplots(1,2, figsize=(12,2))
axes[0].plot(x,y, color = 'purple', ls = '--',lw='3')
axes[1].plot(x,z, color = 'red', ls = '-',lw='3')
```
